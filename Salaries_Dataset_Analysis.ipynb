```python
cleaned_df.head()
```

```python
#top 5 with the lowest risk
low_risk = cleaned_df.sort_values('Spread',ascending=False)
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#top 5 with the highest potential
highest_potental = cleaned_df.sort_values('Mid-Career 90th Percentile Salary',ascending=False)
highest_potental[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
#highest spread
highest_spread = cleaned_df.sort_values('Mid-Career Median Salary',ascending=False)
highest_spread[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```
